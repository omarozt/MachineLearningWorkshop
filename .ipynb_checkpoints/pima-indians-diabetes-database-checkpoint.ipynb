```python
def plot_corr(df,size=11):
    corr=df.corr()
    fig,ax = plt.subplots(figsize=(size,size))
    ax.matshow(corr)
    cax=ax.matshow(corr,interpolation='nearest')
    fig.colorbar(cax)
    plt.xticks(range(len(corr.columns)),corr.columns,rotation = 45)
    plt.yticks(range(len(corr.columns)),corr.columns)
    
```

```python
del pima["skin"]
plot_corr(pima)
```

## Change data from true, false to 1, 0 for diabetes column

```python
diabetes_map = {True:1, False:0}


pima['diabetes']=pima['diabetes'].map(diabetes_map)
pima['diabetes']
```

```python
ndiabetes_true=len(pima.loc[pima['diabetes']==1])
ndiabates_false=len(pima.loc[pima['diabetes']==0])

print('Diyabet olan hasstalarin orani: %{0:2.2f}'.format(ndiabetes_true/(ndiabetes_true+ndiabates_false)))
```

## Train - Test 

```python
#from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split

feature_column_names = pima.columns[:-1]
predicted_class_names = pima.columns[-1:]

x = pima[feature_column_names].values
y = pima[predicted_class_names].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
print ("train data orani : {0}".format(len(X_train)/len(pima)))
```

```python
pima.head()
```

```python
for column in pima.columns:
    print("Number of zeroes in {1:12} : {0:>4}".format(len(pima.loc[pima[column]==0]),column))
```

```python
from sklearn.preprocessing import Imputer
```

```python
X_train[45] ## handpick a row with a zero value
```

```python
fill_0 = Imputer(missing_values=0,strategy='mean',axis=0)
```

```python
X_train = fill_0.fit_transform(X_train)
X_test = fill_0.fit_transform(X_test)
```

```python
X_train[45] ## we've replaced zeros with mean values of test or train data
```

```python
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()
#y_train.ravel()

nb_model.fit(X_train,y_train.ravel())
```

# Performance of Naive Bayes

```python
nb_predict_train = nb_model.predict(X_train)

#performance metric 

from sklearn import metrics

metrics.accuracy_score(y_train,nb_predict_train)
```

```python
nb_predict_test = nb_model.predict(X_test)
metrics.accuracy_score(y_test,nb_predict_test)
```

# Details

```python
conf_mx = metrics.confusion_matrix(y_test,nb_predict_test)
print(conf_mx)
```

```python
classification_report=metrics.classification_report(y_test,nb_predict_test)
print(classification_report)
```

```python
label = ["0","1"]
sns.heatmap(conf_mx, annot=True, xticklabels=label, yticklabels=label)
```
